```python
import geopandas as gpd
```

```python
nationalities = gpd.read_file('Nationalities of migrants in each governorate.csv')
```

```python
nationalities.head()
```

```python
lbn = gpd.read_file('LBN_adm/adm1_centroids.shp')
```

```python
lbn.plot()
```

```python
nationalities = nationalities.rename(columns={'Governorate':'district'})
```

```python
lbn = lbn.rename(columns={'NAME_1':'district'})
```

```python
lbn_shp = lbn.merge(nationalities, on='district')
```

```python
lbn_shp.head()
```

```python
lbn_shp
```

```python
lbn_shp['geometry_x'] = lbn_shp['geometry']
```

```python
lbn_shapes = lbn [['geometry', 'district']]
lbn_shapes.head()
```

```python
nat_loc = nationalities [['nationality_name', 'SUM total_ind', 'district']]
nat_loc
```

```python
lbn_shapes = lbn_shapes.merge(nat_loc, on='district')
lbn_shapes.head()
```

```python
lbn_shapes.plot()
```

```python
lbn_shapes
```

```python
lbn_shapes.to_file('nationalities.shp')
```
